```python
#  Theory + All Commands (Colab Executable)
# Uses Python + SQLite so every command actually executes and shows output.

import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cur = conn.cursor()

def run(query):
    """Executes a query. Returns a table if it's a SELECT, else a status message."""
    cur.execute(query)
    conn.commit()
    if query.strip().upper().startswith("SELECT"):
        cols = [d[0] for d in cur.description]
        return pd.DataFrame(cur.fetchall(), columns=cols)
    return "Query executed successfully."

# SQL (Structured Query Language) is a language used to communicate with databases.
# A database stores information in an organized way.

# Using SQL, we can:
# - Retrieve data, Insert new data,Update existing data, Delete data, Create databases and tables

# RDBMS = Relational Database Management System.
# It is software used to manage relational databases.
# Examples: MySQL, SQL Server, Oracle, PostgreSQL, Microsoft Access.
# All of these support SQL.

# A TABLE stores data in rows and columns.
# A ROW (record) represents one complete entry.
# A COLUMN (field) represents one type of information; each column
# stores similar data (e.g. all Names, all Ages).
# A database can contain many tables.

# SQL SYNTAX
# An SQL statement is a command that tells the database what to do.
# Example: SELECT * FROM Customers; -> retrieves all data from Customers.

# SQL keywords: SELECT, FROM, WHERE, INSERT, UPDATE, DELETE, CREATE, etc.
# These keywords have special, predefined meanings.

# SQL is NOT case sensitive:
#   SELECT * FROM Students;
#   select * from students;
#   SeLeCt * FrOm stUdents;
# All three are valid, but writing keywords in UPPERCASE is good practice.

# A semicolon (;) marks the end of a statement.
# If multiple statements are written together, semicolons separate them:
#   SELECT * FROM Customers;
#   SELECT * FROM Orders;

# Create a table to work with:
run("""
CREATE TABLE Students (
    StudentID INTEGER,
    Name TEXT,
    Age INTEGER,
    Department TEXT,
    City TEXT,
    Email TEXT
);
""")

# Insert some sample rows so all queries below return real results
run("INSERT INTO Students VALUES (1, 'Ali', 20, 'CS', 'Lahore', 'ali@gmail.com');")
run("INSERT INTO Students VALUES (2, 'Sara', 21, 'IT', 'Karachi', NULL);")
run("INSERT INTO Students VALUES (3, 'Ahmed', 19, 'CS', 'Lahore', 'ahmed@gmail.com');")
run("INSERT INTO Students VALUES (4, 'Ayesha', 22, 'IT', 'Sahiwal', 'ayesha@gmail.com');")
run("INSERT INTO Students VALUES (5, 'Zarmeen', 23, 'CS', 'Karachi', NULL);")






## SQL SELECT(The SELECT statement retrieves data from a table.)
# Syntax: SELECT column1, column2 FROM table_name;

# Select specific columns
print(run("SELECT Name, Department FROM Students;"))

# SELECT * -> the asterisk means "all columns"
print(run("SELECT * FROM Students;"))





## SQL SELECT DISTINCT
# Sometimes a column contains duplicate values.
# Without DISTINCT, duplicates are included:
print(run("SELECT Department FROM Students;"))

# With DISTINCT, only unique values are returned:
print(run("SELECT DISTINCT Department FROM Students;"))

# COUNT(DISTINCT column) counts the number of unique values
print(run("SELECT COUNT(DISTINCT Department) AS UniqueDepartments FROM Students;"))

# COUNT(column) counts all non-NULL values (duplicates included)
print(run("SELECT COUNT(Department) AS TotalDepartmentEntries FROM Students;"))





## SQL WHERE( The WHERE clause filters records based on a condition.)
# Syntax:SELECT * FROM table_name WHERE condition;
# Text values must be in single quotes:   WHERE Department = 'CS'
# Numeric values need no quotes:          WHERE Age = 20

print(run("SELECT * FROM Students WHERE Department = 'CS';"))

# Operators used with WHERE:
print(run("SELECT * FROM Students WHERE Age = 20;"))              # Equal
print(run("SELECT * FROM Students WHERE Age > 20;"))              # Greater than
print(run("SELECT * FROM Students WHERE Age < 21;"))              # Less than
print(run("SELECT * FROM Students WHERE Age >= 21;"))             # Greater or equal
print(run("SELECT * FROM Students WHERE Age <= 21;"))             # Less or equal
print(run("SELECT * FROM Students WHERE Department <> 'CS';"))    # Not equal
print(run("SELECT * FROM Students WHERE Age BETWEEN 19 AND 22;")) # Range (inclusive)
print(run("SELECT * FROM Students WHERE Name LIKE 'A%';"))        # Pattern: starts with A
print(run("SELECT * FROM Students WHERE Department IN ('IT', 'CS');"))  # Match multiple values





## SQL ORDER BY
# The ORDER BY keyword sorts the result set.
# By default, SQL sorts in Ascending (ASC) order.

print(run("SELECT * FROM Students ORDER BY Name;"))          # A to Z
print(run("SELECT * FROM Students ORDER BY Name DESC;"))     # Z to A
print(run("SELECT * FROM Students ORDER BY Age DESC;"))      # Highest to Lowest

# ORDER BY multiple columns: sorts by the first column,
# then by the second column when the first column has ties.
print(run("SELECT * FROM Students ORDER BY Department, Name;"))

# Combine ASC and DESC
print(run("SELECT * FROM Students ORDER BY Department ASC, Name DESC;"))





## SQL AND
# The AND operator combines multiple conditions.
# A record is returned only if ALL conditions are true.
print(run("SELECT * FROM Students WHERE Department = 'CS' AND City = 'Lahore';"))
print(run("SELECT * FROM Students WHERE City = 'Lahore' AND Name LIKE 'A%';"))





## SQL OR
# The OR operator returns rows if AT LEAST ONE condition is true.
print(run("SELECT * FROM Students WHERE City = 'Lahore' OR Department = 'IT';"))
print(run("SELECT * FROM Students WHERE Department IN ('CS', 'IT') OR Name LIKE 'A%';"))

# Combining AND and OR: always use parentheses to make the logic clear.
print(run("""
SELECT * FROM Students
WHERE Department = 'CS'
AND (Name LIKE 'A%' OR Name LIKE 'Z%');
"""))
# Without parentheses, SQL may interpret the logic differently and
# return unintended rows -- always use parentheses when mixing AND/OR.




## SQL NOT
# The NOT operator reverses a condition: "give me everything except.."
print(run("SELECT * FROM Students WHERE NOT Department = 'CS';"))         # exclude CS
print(run("SELECT * FROM Students WHERE Name NOT LIKE 'A%';"))            # names not starting with A
print(run("SELECT * FROM Students WHERE Age NOT BETWEEN 20 AND 22;"))     # outside the range
print(run("SELECT * FROM Students WHERE Department NOT IN ('IT');"))      # exclude IT
print(run("SELECT * FROM Students WHERE NOT Age > 21;"))                  # means Age <= 21
print(run("SELECT * FROM Students WHERE NOT Age < 21;"))                  # means Age >= 21



## SQL INSERT INTO
# The INSERT INTO statement adds new records to a table.

# Insert all columns
run("INSERT INTO Students VALUES (6, 'Bilal', 24, 'IT', 'Multan', 'bilal@gmail.com');")
# Insert specific columns only -- remaining columns become NULL/default
run("INSERT INTO Students (Name, Age) VALUES ('Hina', 20);")
# Insert multiple rows at once
run("""
INSERT INTO Students (StudentID, Name, Age)
VALUES
(8, 'Usman', 21),
(9, 'Fatima', 20),
(10, 'Noor', 19);
""")
print(run("SELECT * FROM Students;"))



## SQL NULL VALUES
# A NULL value means no value or missing data.
# It does NOT mean 0 (zero) or '' (empty string) -- it means the
# value is unknown or was never provided.

# Wrong way to check for NULL (always returns nothing):
#   SELECT * FROM Students WHERE Email = NULL;

# IS NULL finds records where a column has no value
print(run("SELECT * FROM Students WHERE Email IS NULL;"))
# IS NOT NULL finds records that do contain a value
print(run("SELECT * FROM Students WHERE Email IS NOT NULL;"))







## SQL UPDATE
# The UPDATE statement modifies existing data in a table.
# Syntax:UPDATE table_name SET column1 = value1, column2 = value2 WHERE condition;

run("UPDATE Students SET Department = 'CS' WHERE Name = 'Bilal';")
# Update multiple columns at once
run("UPDATE Students SET Name = 'Alisha', City = 'Sahiwal' WHERE StudentID = 1;")
# Update multiple rows (every row matching the condition)
run("UPDATE Students SET Department = 'IT' WHERE City = 'Lahore';")
print(run("SELECT * FROM Students;"))

# IMPORTANT WARNING: if the WHERE clause is forgotten, ALL rows are updated:
#   UPDATE Students SET City = 'Lahore';
# This would change the City of every single student always double-check WHERE.




## SQL DELETE
# DELETE removes records (rows) from a table.
# Syntax:DELETE FROM table_name WHERE condition;

run("DELETE FROM Students WHERE StudentID = 10;")
# Delete multiple rows
run("DELETE FROM Students WHERE Department = 'CS' AND Age < 20;")
print(run("SELECT * FROM Students;"))

# Delete ALL records (table structure stays, but becomes empty)
#   DELETE FROM Students;
# DELETE vs DROP TABLE:
#   DELETE FROM Students;  -> removes all rows, table still exists
#   DROP TABLE Students;   -> removes the table, its data AND structure completely





## SQL SELECT TOP
# These keywords limit the number of rows returned.
# Database    | Keyword
# SQL Server  | TOP
# MySQL/SQLite| LIMIT
# Oracle      | FETCH FIRST

print(run("SELECT * FROM Students LIMIT 3;"))                       # first 3 rows
# LIMIT with WHERE
print(run("SELECT * FROM Students WHERE Department = 'IT' LIMIT 2;"))
# LIMIT with ORDER BY
print(run("SELECT * FROM Students ORDER BY Age DESC LIMIT 3;"))     # 3 oldest students




## SQL AGGREGATE FUNCTIONS
# Aggregate functions perform a calculation across multiple rows
# and return a single result. They ignore NULL values (except COUNT(*),
# which counts every row regardless of NULLs).

print(run("SELECT MIN(Age) AS Youngest FROM Students;"))        # smallest value
print(run("SELECT MAX(Age) AS Oldest FROM Students;"))          # largest value
print(run("SELECT COUNT(*) AS TotalStudents FROM Students;"))   # counts all rows
print(run("SELECT COUNT(*) AS CSStudents FROM Students WHERE Department = 'CS';"))
print(run("SELECT SUM(Age) AS TotalAge FROM Students;"))        # adds all values
print(run("SELECT AVG(Age) AS AverageAge FROM Students;"))      # average value
conn.close()


# Keyword           Purpose                         Example

# SELECT            Retrieve data                    SELECT * FROM Students;
# DISTINCT          Unique values only                SELECT DISTINCT Department FROM Students;
# WHERE             Filter records                    WHERE Department = 'CS';
# ORDER BY          Sort results                      ORDER BY Age DESC;
# AND               All conditions true               WHERE Age>18 AND City='Lahore';
# OR                Any one condition true            WHERE City='Lahore' OR City='Karachi';
# NOT               Reverse a condition                WHERE NOT Department='CS';
# INSERT INTO       Add new rows                      INSERT INTO Students VALUES (...);
# IS NULL           Find missing values                WHERE Email IS NULL;
# IS NOT NULL       Find filled values                  WHERE Email IS NOT NULL;
# UPDATE            Modify existing rows               UPDATE Students SET City='Lahore' WHERE StudentID=1;
# DELETE            Remove rows                        DELETE FROM Students WHERE StudentID=1;
# DROP TABLE        Delete whole table                 DROP TABLE Students;
# LIMIT/TOP/FETCH   Limit number of rows               LIMIT 5 / TOP 5 / FETCH FIRST 5 ROWS ONLY
# MIN() / MAX()     Smallest / Largest value           MIN(Age) / MAX(Age)
# COUNT()           Count rows                         COUNT(*)
# SUM() / AVG()     Total / Average                    SUM(Age) / AVG(Age)
```
